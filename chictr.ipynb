```python
import requests
import re
```

```python
hdf_url = 'http://www.chictr.org.cn/showproj.aspx?proj=16624'
try:
    headers={
            "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0"
            }
    hdf_html=requests.get(hdf_url,headers=headers)
    hdf_html.raise_for_status()  # 如果 HTTP 请求返回了不成功的状态码， Response.raise_for_status() 会抛出一个 HTTPError 异常。
#     hdf_html.encoding = r.apparent_encoding
    print(hdf_html.encoding)
    print(hdf_html.content.decode('utf-8'))
#     print(hdf_html.headers)

except:
    print("爬取失败！")
```

```python
rule_register_code = r'Registration number：.*?<td>.*?(.*?)&nbsp;'
rule_applyconnector_name = r'申请注册联系人：.*?"cn">(.*?)&nbsp'
rule_studyconnector_name = r'研究负责人：.*?"cn">(.*?)&nbsp'
rule_tel = r'Applicant telephone：.*?">(.*?)&nbsp'
rule_aemail = r'Applicant E-mail：.*?">(.*?)&nbsp'
rule_semail = r's E-mail：.*?<td>(.*?)&nbsp'
rule_prov = r'Secondary sponsor：.*?省\(直辖市\)：.*?"cn">(.*?)</p>'
rule_city = r'Secondary sponsor：.*?市\(区县\)：</p>.*?"cn">(.*?)</p>'
rule_list1 = [rule_register_code,rule_applyconnector_name,rule_tel,rule_aemail,rule_prov,rule_city]
rule_list2 = [rule_register_code,rule_studyconnector_name,rule_tel,rule_semail,rule_prov,rule_city]
pattern = re.compile(rule_city, re.S)
results = re.findall(pattern, hdf_html.text)
print(results)
print(type(results[0].strip()))
```

```python
def get_testdata(url_list):
    res = []
    for url in url_list:
        try:
            headers={
                    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0"
                    }
            hdf_html=requests.get(url,headers=headers)
            hdf_html.raise_for_status()  # 如果 HTTP 请求返回了不成功的状态码， Response.raise_for_status() 会抛出一个 HTTPError 异常。
            # hdf_html.encoding = r.apparent_encoding
            # print(hdf_html.encoding)
            # print(hdf_html.text)

        except:
            print("爬取失败！")
        # 获取apply_connector的测试信息
        applyconnector=[]
        for rule in rule_list1:
            pattern = re.compile(rule, re.S)
            results = re.findall(pattern, hdf_html.text)
            applyconnector.append(results[0].strip())
        applyconnector.append('apply_connector')
        res.append(tuple(applyconnector))
        # 获取study_connector的测试信息
        studyconnector=[]
        for rule in rule_list2:
            pattern = re.compile(rule, re.S)
            results = re.findall(pattern, hdf_html.text)
            studyconnector.append(results[0].strip())
        studyconnector.append('study_connector')
        res.append(tuple(studyconnector))
    
    return res
```

```python
url_list = ['http://www.chictr.org.cn/showproj.aspx?proj=16624','http://www.chictr.org.cn/showproj.aspx?proj=30682']
get_testdata(url_list)
```
