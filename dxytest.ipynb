```python
import requests
import re
```

```python
dxy_url = 'https://dxy.com/health/hospital/19129'
try:
    headers={
            "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0"
            }
    dxy_html=requests.get(dxy_url,headers=headers)
    dxy_html.raise_for_status()  # 如果 HTTP 请求返回了不成功的状态码， Response.raise_for_status() 会抛出一个 HTTPError 异常。
#     hdf_html.encoding = r.apparent_encoding
    print(dxy_html.encoding)
    print(dxy_html.content.decode('utf-8'))
#     print(hdf_html.headers)

except:
    print("爬取失败！")
```

```python
hospital_id = re.findall(r'.*?hospital/(\w+)', dxy_url)[0]
print(hospital_id)

def getcity(address):
    rule = u'(.*?)市'
    pattern = re.compile(rule, re.S)
    results = re.findall(pattern, address)[0]
    
    return results+u'市'
```
